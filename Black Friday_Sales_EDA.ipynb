**1. Purchase Behavior Analysis:**
Analyze the distribution of the Purchase amount. Are there specific groups (e.g., age groups, gender, or city categories) that tend to spend more?


```python
distribution_gender = df.groupby('Gender')['Purchase'].sum()
distribution_gender
```

```python
distribution_gender.plot(kind="pie",label="Purchase",autopct='%1.1f%%')
plt.title(" Purchase Distribution by Gender")
plt.legend(['Gender'])
plt.show()
```

```python
distribution_city = df.groupby('City_Category')['Purchase'].sum()
distribution_city
```

```python
distribution_city.plot(kind="pie",label="Purchase",autopct='%.1f%%')
plt.title(" Purchase Distribution by City")
plt.legend(['City'])
plt.show()
```

```python
Age_grouping = df.groupby('Age')['Purchase'].sum().reset_index()
Age_grouping
```

```python
sns.barplot(x='Age', y='Purchase', data=Age_grouping)
plt.title('Purchase by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Total Purchase')
plt.xticks(rotation=45)
plt.show()
```

After analysing this case,

- Purchase amount of male are higher then the female.
- Purchase distribution of city category B is higher followed by city C and A.
- People of age group 26-35 having higher purchase rate.

**Customer Demographics:**

3. Analyze the distribution of customers based on Age, Gender, and City_Category. Which demographic group is most active during the Black Friday sales?

```python
Age_distribution = df.groupby('Age')['User_ID'].count().reset_index().plot(kind="bar",color="green",edgecolor="black",x="Age",y="User_ID")
plt.title(" Customer Distribution by Age")
plt.xlabel("Age")
plt.ylabel("Total Customers")
for i in Age_distribution.containers:
    Age_distribution.bar_label(i)
plt.show()
```

```python
Gender_distribution = df.groupby('Gender')['User_ID'].count().reset_index().plot(kind="bar",color="red",edgecolor="black",x="Gender",y="User_ID")
plt.title(" Customer Distribution by Gender")
plt.xlabel("Gender")
plt.ylabel("Total Customers")
for i in Gender_distribution.containers:
    Gender_distribution.bar_label(i)
plt.show()
```

```python
City_distribution = df.groupby('City_Category')['User_ID'].count().reset_index().plot(kind="bar",color="blue",edgecolor="black",x="City_Category",y="User_ID")
plt.title(" Customer Distribution by City")
plt.xlabel("City_Category")
plt.ylabel("Total Customers")
for i in City_distribution.containers:
    City_distribution.bar_label(i)
plt.show()
```

After analysing these cases,

- Age group of 26-35 Customer are maximum.
- Male customers are having maximum number.
- City B is having highest count.

**Stay in Current City**

6. Analyze how the duration of stay in the current city (Stay_In_Current_City_Years) affects purchase behavior. Do longer stays correlate with higher spending?

**Correlation Between Product Categories:**

7. Examine if there's any correlation between the three product categories (Product_Category_1, Product_Category_2, and Product_Category_3). Do customers who buy from one category tend to buy from another?


```python
Product_Categories = df[['Product_Category_1', 'Product_Category_2', 'Product_Category_3']]
Product_Categories
```

```python
Product_Categories.corr()
```

```python
sns.heatmap(Product_Categories.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
# A correlation value  of 1 indicates a perfect positive correlation between three categories . It says that The person who is buying from 1 tend to buy from other.
```
